```python
# axis
a = np.array([1, 2, 3])
b = np.array([1, 2, 3])

np.sum([a, b], axis=0) # [(1 + 1), (2 + 2), (3, 3)]
np.sum([a, b], axis=1) # [(1 + 2 + 3), (1 + 2 + 3)]
```
