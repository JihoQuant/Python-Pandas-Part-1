### EDA - Exploratory Data Analysis

- In statistics, exploratory data analysis(EDA) is an approach to analyzing data sets to summarize thier main characteristics, often with visual methods
- Two Parts
  - Metadata: data about data
    - 데이터 크기
    - 컬럼명
    - 데이터 타입
    - 비어 있는 데이터
    - etc.
  - Univariate descriptive statistics: summary statistics about individual variables(colums)

### Scalar Value

```python
# at, iat // 상수 접근 시 사용함
df.loc[100, '순이익률(%)']
df.at[100, '순이익률(%)']
```

```python
# DataFrame이 커질수록 loc 보다는 at의 성능이 뛰어나다.
%timeit df.loc[100, '순이익률(%)']
print('-----')
%timeit df.at[100, '순이익률(%)']
```
